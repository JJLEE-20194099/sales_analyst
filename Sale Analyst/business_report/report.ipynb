```python
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):           
        filepath = path + file
        frames.append(pd.read_csv(filepath))


frames
```

Merge dataframe pandas

```python
result = pd.concat(frames)
result
```

1. What was the best month for sales?
   How much was earned that month?

```python
df['Month'] = df['Order Date'].str[:2]
set(df['Month'])
```

Clean data

```python
df = df.dropna(how='all')
df = df[df['Month'] != 'Or']
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast='integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast='float')
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
moving_column = df.pop('Sales')
df.insert(4, 'Sales', moving_column)
df.head()
```

Sales per month

```python
sales_value = df.groupby('Month').sum()['Sales']
type(sales_value)
```

Visualize sales per month bar chart 

```python
months = range(1, 13)
plt.bar(months, sales_value)
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales in USD')
```

2. What city has the best sales?

```python
def address_to_city(sample_address):
    return sample_address.split(',')[1]
```

```python
df['City'] = df['Purchase Address'].apply(address_to_city)
df.head()
```

```python
city_sales_value = df.groupby('City').sum()['Sales']
city_sales_value.max()
```

Visualize sales per city

```python
cities = [city for city, sales in city_sales_value.items()]
plt.bar(cities, city_sales_value)
plt.xticks(cities, rotation=45, size=8)
plt.xlabel('Cities')
plt.ylabel('Sales in USD')
```

3. What time should we display ads to maximize the likelihood of customer's buying product

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Hours'] = df['Order Date'].dt.hour
```

```python
hour_orders_total = df.groupby('Hours').sum()['Quantity Ordered']
hour_orders_total.max()
```

Visualize hour sales bar chart

```python
hours = [hour for hour, sales in hour_orders_total.items()]
plt.plot(hours, hour_orders_total)
plt.grid()
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Sales in USD')
```

Conclusion: We should display ads at 6:30 pm and 11:30 am to maximize the likelihood of customer's buying product

4. What products are most often sold together?

```python
groupProduct = lambda products: ','.join(products)
```

```python
df_dup = df[df['Order ID'].duplicated(keep = False)]
```

```python
together_product = df_dup.groupby('Order ID')['Product'].transform(groupProduct)
```

```python
df_dup['All Products'] = together_product
df_dup = df_dup[['Order ID', 'All Products']].drop_duplicates()
```

Products are most often sold together

5. What product sold the most? Why do you think it sold the most

Get products_ls, sold_per_product_quantity

```python
sold_per_product_quantity = df.groupby('Product').sum()['Quantity Ordered']
products_ls = [product for product, quantity in sold_per_product_quantity.items()]
plt.bar(products_ls, sold_per_product_quantity)
plt.xticks(products_ls, rotation=90, size=8)
plt.xlabel('Products')
plt.ylabel('Quantity')
```

```python
mean_product_prices = df.groupby('Product').mean()['Price Each']
products_ls = [product for product, quantity in mean_product_prices.items()]
plt.bar(products_ls, mean_product_prices)
plt.xticks(products_ls, rotation=90, size=8)
plt.xlabel('Products')
plt.ylabel('Prices In USD')
```

Stacking two bar charts

```python
x = products_ls
y1 = sold_per_product_quantity
y2 = mean_product_prices

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(x, y1, color='g')
ax2.plot(x, y2, 'b-')

ax1.set_xticklabels(products_ls, rotation=90, size=10)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity', color='g')
ax2.set_ylabel('Average Prices', color='b')
```
